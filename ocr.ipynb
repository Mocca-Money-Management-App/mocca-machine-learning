```python
!curl -o receipts.jpg https://res.cloudinary.com/ewik/image/upload/v1683515031/WhatsApp_Image_2023-05-06_at_12.25.33_c34paf.jpg
import cv2
import pytesseract
img = cv2.imread('receipts.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 3)
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

custom_config = r'--oem 3 --psm 6'
text = pytesseract.image_to_string(gray, lang='ind', config=custom_config)
print(text)
```
